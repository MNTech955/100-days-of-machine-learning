What is Univariate Analysis?

Univariate Analysis means analyzing one column/one variable at a time

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('sample_data/train.csv')
df.head()
```

1. Categorical Data

```python

sns.countplot(data=df, x='Survived')
```

```python
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(data=df, x='Pclass', hue='Pclass', palette='viridis',legend=False)
plt.show()
```

```python
sns.countplot(data=df, x='Sex', hue='Sex', palette='viridis',legend=False)
plt.show()
```

```python
sns.countplot(data=df, x='Embarked', hue='Embarked', palette='viridis',legend=False)
plt.show()
```

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%0.1f%%')

plt.show()
```

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%0.1f%%')

plt.show()
```

2. Numerical Data

a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=5)
```

b. Distplot

```python
sns.distplot(df['Age'])
```

c. Boxplot


```python
sns.boxplot(data=df, x='Age')
plt.show()
```
